We import the detection logic from the `src.watermark_module` package.

## 2. Run Analysis on One Image

## 3. Batch Test a Folder

```python
batch_results = analyse_folder("../data/sample_images/watermark")

for result in batch_results:
    if "error" in result:
        print(f"\nError in {result['image_path']}: {result['error']}")
        continue
        
    print()
    print(f"File: {result['image_path']}")
    print("watermark detected: ", result['overall_detectable_watermark_evidence'])
    print("findings: ", result['summary'])
    print("rationale: ")
    
    for item in result['results']:
        print(f"\t - Detected {item['detector_name']}\n\t\t > {item['notes']}")
```
